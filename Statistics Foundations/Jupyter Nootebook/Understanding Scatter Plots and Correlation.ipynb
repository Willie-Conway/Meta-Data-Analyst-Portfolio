# Understanding Scatter Plots and Correlation

![Scatterplot Correlation](Images/Scatterplot_Correlation.png)

### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")
```

### Generate Sample Data

```python
# Generate sample data for babies
np.random.seed(42)
n_samples = 100

# Randomly generate ages, weights, and heights
ages = np.random.randint(0, 12, n_samples)  # Ages from 0 to 11 months
weights = ages * 2.5 + np.random.normal(0, 3, n_samples)  # Weight in kg with some noise
heights = ages * 5 + np.random.normal(0, 10, n_samples)  # Height in cm with some noise

# Create a DataFrame
babies_data = pd.DataFrame({
    'Age (Months)': ages,
    'Weight (kg)': weights,
    'Height (cm)': heights
})

babies_data.head()
```

### Define Types of Correlation

```python
correlation_types = {
    'Positive Correlation': 'As one variable increases, the other variable tends to increase.',
    'Negative Correlation': 'As one variable increases, the other variable tends to decrease.',
    'No Correlation': 'No apparent relationship between the variables.'
}

for key, value in correlation_types.items():
    print(f"{key}: {value}")
```

### Positive Correlation Example

```python
# Scatter plot for positive correlation (Age vs. Height)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Age (Months)', y='Height (cm)', data=babies_data)
plt.title('Scatter Plot: Age vs. Height (Positive Correlation)')
plt.xlabel('Age (Months)')
plt.ylabel('Height (cm)')
plt.show()

# Scatter plot with linear regression
plt.figure(figsize=(10, 5))
sns.regplot(x='Age (Months)', y='Height (cm)', data=babies_data)
plt.title('Scatter Plot with Linear Regression: Age vs. Height')
plt.xlabel('Age (Months)')
plt.ylabel('Height (cm)')
plt.show()
```

### Negative Correlation Example

```python
# Create a negative correlation by reversing weight as age increases
babies_data['Weight (kg)'] = 30 - ages + np.random.normal(0, 2, n_samples)

# Scatter plot for negative correlation (Age vs. Weight)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Age (Months)', y='Weight (kg)', data=babies_data)
plt.title('Scatter Plot: Age vs. Weight (Negative Correlation)')
plt.xlabel('Age (Months)')
plt.ylabel('Weight (kg)')
plt.show()

# Scatter plot with linear regression
plt.figure(figsize=(10, 5))
sns.regplot(x='Age (Months)', y='Weight (kg)', data=babies_data)
plt.title('Scatter Plot with Linear Regression: Age vs. Weight')
plt.xlabel('Age (Months)')
plt.ylabel('Weight (kg)')
plt.show()
```

### No Correlation Example

```python
# Create random data for no correlation
np.random.seed(0)
babies_data['Random'] = np.random.normal(20, 5, n_samples)  # Random variable

# Scatter plot for no correlation (Random vs. Weight)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Random', y='Weight (kg)', data=babies_data)
plt.title('Scatter Plot: Random vs. Weight (No Correlation)')
plt.xlabel('Random Variable')
plt.ylabel('Weight (kg)')
plt.show()

# Scatter plot with linear regression
plt.figure(figsize=(10, 5))
sns.regplot(x='Random', y='Weight (kg)', data=babies_data)
plt.title('Scatter Plot with Linear Regression: Random vs. Weight')
plt.xlabel('Random Variable')
plt.ylabel('Weight (kg)')
plt.show()
```
