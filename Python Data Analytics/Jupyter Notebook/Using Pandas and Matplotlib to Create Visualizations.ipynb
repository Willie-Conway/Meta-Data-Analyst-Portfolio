### Scatter Plot

```python
df.plot.scatter(x='total_bill', y='tip')
```

```python
df.plot.scatter(x='total_bill', y='tip', alpha=0.25)
```

```python
df.plot.scatter(x='size', y='tip', alpha=0.25)
df.plot.scatter(x='size', y='total_bill', alpha=0.25)
```

```python
ax = df.plot.scatter(x='size', y='tip', alpha=0.1, label='tip')
ax = df.plot.scatter(x='size', y='total_bill', alpha=0.1, color='red', label='total_bill', ax=ax)
ax.set_ylabel('Dollars')
```
