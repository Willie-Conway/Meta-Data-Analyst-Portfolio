### Load Data

```python
import pandas as pd
```

```python
# https://rdrr.io/cran/reshape2/man/tips.html
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/tips.csv')
```

```python
df.info()
df.head()
```

### Let's Explore with More Visualizations!

#### Scatter Plot

```python
df.plot.scatter(x='size', y='tip', alpha=0.1)
df.plot.scatter(x='size', y='total_bill', color='red', alpha=0.1)
```

```python
ax = df.plot.scatter(x='size', y='tip', alpha=0.1, label='tip',)
```

```python
ax
```

```python
ax = df.plot.scatter(x='size', y='tip', alpha=0.1, label='tip',)
ax = df.plot.scatter(x='size', y='total_bill', alpha=0.1, color='red', label='total_bill', ax=ax,)
```

```python
ax = df.plot.scatter(x='size', y='tip', alpha=0.1, label='tip',)
ax = df.plot.scatter(x='size', y='total_bill', alpha=0.1, color='red', label='total_bill', ax=ax,)
ax.set_ylabel('Dollars')
```

```python
ax = df.plot.scatter(x='size', y='tip', alpha=0.1, label='tip',)
ax = df.plot.scatter(x='size', y='total_bill', alpha=0.1, color='red', label='total_bill', ax=ax,)
ax.set_ylabel('Dollars')
ax.set_title('Party Size Bill & Tips')
```

```python
ax = df.plot.scatter(x='size', y='tip', alpha=0.1, label='tip',)
ax = df.plot.scatter(x='size', y='total_bill', alpha=0.1, color='red', label='total_bill', ax=ax,)
plt.ylabel('Dollars')
plt.title('Party Size Bill & Tips')
```

#### New Feature

```python
df['tip_percent'] = df.tip / df.total_bill
```

```python
df.tip_percent.describe()
```

```python
ax = df.plot.box(column='tip_percent', vert=False)
```

```python
df.plot.scatter(x='total_bill', y='tip_percent', alpha=0.25, color='green')
```

```python
df.plot.hist(column='tip_percent',)
```

```python
df.plot.hist(column='tip_percent', bins=50)
```

```python
df.plot.hist(column='tip_percent', bins=50, range=(0,1))
```
