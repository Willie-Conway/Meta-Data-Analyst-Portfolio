```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# https://rdrr.io/cran/reshape2/man/ips.html
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/tips.csv')
```

```python
df.info()
df.head()
```

```python
days_ordered = ['Thur', 'Fri', 'Sat', 'Sun']
df['day'] = pd.Categorical(df['day'], categories = days_ordered, ordered=True)
```

### Explore

### Day

```python
sns.barplot(data=df, x='day', y='total_bill', errorbar=None, estimator='median')
```

```python
sns.barplot(data=df, x='day', y='total_bill', errorbar=None, estimator='sum')
```

```python
sns.barplot(data=df, x='day', y='total_bill', errorbar=None, estimator='count')
plt.ylabel('Count')
```

### Party Size

```python
sns.boxplot(data=df, x='size', y='total_bill',)
```

```python
sns.barplot(data=df, x='day', y='total_bill', hue='size', errorbar=None, estimator='count')
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', alpha=0.7, hue='size')
```

### Explanatory

```python
ax = sns.barplot(data=df, x='day', y='total_bill', errorbar=None, estimator='median',
                palette=['grey', 'grey', 'red', 'red',],
                )
ax.set_title('Weekends Tend to Have Largest Bill')
ax.set_ylabel('Total bill (Median)')
ax.set_xlabel('Day of the Week')
```

```python
ax = sns.barplot(
    data=df, x='size', y='total_bill', estimator='median', errorbar=None,
    color='green',
)
ax.set_title('Total Bill Increases w/ Party Size')
ax.set_ylabel('Total Bill (Median)')
ax.set_xlabel('Party size')
```

```python
ax = sns.barplot(
    data=df, x='size', y='total_bill', errorbar=None, estimator='count',
    palette=['grey', 'red', 'grey', 'grey', 'grey', 'grey',]
)
ax.set_title('Most Common Party Size: 2 People')
ax.set_ylabel('Count')
ax.set_xlabel('Party Size')
```

```python
ax = sns.barplot(
    data=df, x='day', y='total_bill', hue='size', errorbar=None, estimator='count',
    palette=['grey', 'grey', 'red', 'red', 'grey', 'grey',]
)
ax.set_title('Party Sizes of 3 or 4 Most Common on Weekends')
ax.set_ylabel('Count')
ax.set_xlabel('Day of the Week')
```

```python
sns.scatterplot(data=df, x='total_bill', y= 'tip', alpha=0.7, hue='size')
```

```python
party_subset = df[df['size'].isin([2,3,4])]
```

```python
sns.scatterplot(data=party_subset, x='total_bill', y='tip', hue='size')
```

```python
sns.lmplot(
    data= party_subset,
    x='total_bill',
    y='tip',
    hue='size',   
)
```

```python
sns.lmplot(
    data= party_subset,
    x='total_bill',
    y='tip',
    hue='size',
    scatter=False,
)
```

```python
sns.lmplot(
    data= party_subset,
    x='total_bill',
    y='tip',
    hue='size',
    scatter=False,
    ci=None,
)
```
