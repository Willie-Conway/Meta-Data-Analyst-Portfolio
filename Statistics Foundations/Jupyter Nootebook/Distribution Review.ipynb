### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Set random seed for reproducibility

```python
np.random.seed(42)
```

### Function to generate customer age data

```python
def generate_customer_ages(size):
    """Generate a dataset of customer ages between 18 and 70."""
    return np.random.randint(18, 70, size=size)  # Random ages between 18 and 70
```

### Generate Customer Age Data

```python
customer_ages = generate_customer_ages(100)
```

### Save to CSV

```python
customer_age_df = pd.DataFrame({"Customer Age": customer_ages})
customer_age_df.to_csv("csv/Customers_Age.csv", index=False)
```

### Function to plot histograms

```python
def plot_histogram(data, title, xlabel):
    """Plot a histogram with kernel density estimate."""
    plt.figure(figsize=(10, 6))
    sns.histplot(data, bins=10, kde=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.grid()
    plt.show()
```

## Visualizing Different Distributions

### Normal Distribution

```python
normal_data = np.random.normal(loc=35, scale=10, size=1000)  # Mean age = 35, std dev = 10
plot_histogram(normal_data, 'Normal Distribution of Ages', 'Age')
```

### Uniform Distribution

```python
uniform_data = np.random.uniform(low=18, high=70, size=1000)  # Age between 18 and 70
plot_histogram(uniform_data, 'Uniform Distribution of Ages', 'Age')
```

### Poisson Distribution

```python
# Using a Poisson distribution with lambda = 35 to simulate age occurrences
poisson_data = np.random.poisson(lam=35, size=1000)  # Mean = 35
plot_histogram(poisson_data, 'Poisson Distribution of Ages', 'Age')
```

### Exponential Distribution

```python
exponential_data = np.random.exponential(scale=10, size=1000)  # Scale parameter (mean) = 10
plot_histogram(exponential_data, 'Exponential Distribution of Ages', 'Age')
```
