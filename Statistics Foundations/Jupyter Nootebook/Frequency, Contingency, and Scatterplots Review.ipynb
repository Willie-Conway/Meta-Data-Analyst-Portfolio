# Frequency, Contingency, and Scatterplots Review

### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")
```

### Create Sample Data

```python
# Sample data for customer subscriptions
data = {
    'Subscriber Status': np.random.choice(['Active', 'Expired', 'Nonsubscriber'], size=500),
    'Gender': np.random.choice(['Male', 'Female'], size=500),
    'Age Group': np.random.choice(['<21', '21-30', '31-40', '41+'], size=500)
}

df = pd.DataFrame(data)
df.head()
```

### Frequency Table

```python
# Frequency table for Subscriber Status
frequency_table = df['Subscriber Status'].value_counts()
frequency_table = frequency_table.reset_index()
frequency_table.columns = ['Subscriber Status', 'Count']

print("Frequency Table for Subscriber Status:")
print(frequency_table)
```

### Visualize Frequency Table

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='Subscriber Status', y='Count', data=frequency_table)
plt.title('Frequency of Subscriber Status')
plt.xlabel('Subscriber Status')
plt.ylabel('Count')
plt.show()
```

### Contingency Table

```python
# Contingency table for Subscriber Status and Gender
contingency_table = pd.crosstab(df['Subscriber Status'], df['Gender'])
print("Contingency Table for Subscriber Status and Gender:")
print(contingency_table)
```

### Visualize Contingency Table

```python
contingency_table.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('Contingency Table: Subscriber Status by Gender')
plt.xlabel('Subscriber Status')
plt.ylabel('Count')
plt.show()
```

### Correlation and Scatter Plots

```python
# Create sample data for correlation analysis
purchases = np.random.randint(1, 20, size=100)  # Number of purchases
spendings = 100 - (purchases * 5) + np.random.normal(0, 5, size=100)  # Spending amount

correlation_data = pd.DataFrame({
    'Number of Purchases': purchases,
    'Amount Spent': spendings
})

# Calculate correlation coefficient
correlation_coefficient = correlation_data['Number of Purchases'].corr(correlation_data['Amount Spent'])
print(f"Correlation Coefficient: {correlation_coefficient:.2f}")
```

### Scatter Plot with Linear Regression

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Number of Purchases', y='Amount Spent', data=correlation_data)
plt.title('Scatter Plot: Number of Purchases vs. Amount Spent')
plt.xlabel('Number of Purchases')
plt.ylabel('Amount Spent')
plt.show()
```

### Interpreting Correlation

```python
if correlation_coefficient > 0.7:
    print("There is a strong positive correlation.")
elif correlation_coefficient < -0.7:
    print("There is a strong negative correlation.")
elif correlation_coefficient > 0.3:
    print("There is a moderate positive correlation.")
elif correlation_coefficient < -0.3:
    print("There is a moderate negative correlation.")
else:
    print("There is little to no correlation.")
```
