```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statistics import mean, median, mode, StatisticsError
import pandas as pd
import random
```

```python
# Step 1: Create the CSV file with 100 entries
home_prices = [random.randint(100000, 500000) for _ in range(100)]

df = pd.DataFrame(home_prices, columns=['Price'])
df.to_csv('csv/home_prices.csv', index=False)

print(df)
```

```python
# Step 2: Load the dataset from CSV
df = pd.read_csv('csv/home_prices.csv')
home_prices = df['Price'].tolist()  # Convert the DataFrame column to a list

print(home_prices)
```

```python
# Step 3: Calculate Mean
mean_value = mean(home_prices)
print(mean_value)
```

```python
# Step 4: Calculate Median
sorted_prices = sorted(home_prices)
n = len(sorted_prices)
median_value = sorted_prices[n // 2] if n % 2 == 1 else (sorted_prices[n // 2 - 1] + sorted_prices[n // 2]) / 2

print(sorted_prices)
print(n)
print(median_value)
```

```python
# Step 5: Calculate Mode
try:
    mode_value = mode(home_prices)
except StatisticsError:
    mode_value = "No unique mode"
```

```python
# Step 6: Print results
print("Dataset (first 10 entries):", home_prices[:10])  # Show only first 10 for brevity
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
# Step 7: Visualizations

# Set the figure size larger
plt.figure(figsize=(14, 10))  # Adjust width and height as needed

# Scatter Plot
plt.subplot(231)
plt.scatter(range(len(home_prices)), home_prices, color='blue')
plt.title('Scatter Plot of Home Prices')
plt.xlabel('Index')
plt.ylabel('Home Price')
```

```python
# Set the figure size larger
plt.figure(figsize=(14, 10))  # Adjust width and height as needed

# Bar Graph
plt.subplot(232)
plt.bar(range(len(home_prices)), home_prices, color='orange')
plt.title('Bar Graph of Home Prices')

# Show only every 10th house on x-ticks
plt.xticks(range(0, len(home_prices), 10), [f'House {i+1}' for i in range(0, len(home_prices), 10)], rotation=45)  
plt.ylabel('Home Price')
plt.xlabel('House Index')  # Add x-axis label here
```

```python
# Set the figure size larger
plt.figure(figsize=(14, 10))  # Adjust width and height as needed

# Linear Regression Graph
plt.subplot(233)
x = np.array(range(len(home_prices))).reshape(-1, 1)
y = np.array(home_prices)
sns.regplot(x=x.flatten(), y=y, marker='o', color='green')
plt.title('Linear Regression of Home Prices')
plt.xlabel('Index')
plt.ylabel('Home Price')
```

```python
# Set the figure size larger
plt.figure(figsize=(14, 10))  # Adjust width and height as needed

# Pie Chart (limited to the first 10 entries)
plt.subplot(234)
plt.pie(home_prices[:10], labels=[f'House {i+1}' for i in range(10)], autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Home Prices (First 10)')
```

```python
# Set the figure size larger
plt.figure(figsize=(14, 10))  # Adjust width and height as needed

# Histogram
plt.subplot(235)
plt.hist(home_prices, bins=10, color='purple', alpha=0.7)  # Adjust number of bins for clarity
plt.title('Histogram of Home Prices')
plt.xlabel('Home Price')
plt.ylabel('Frequency')

# Set custom x-tick labels and rotate them
plt.xticks(np.arange(100000, 500001, 50000), rotation=45)  # Show ticks at intervals of 50,000 and rotate

plt.tight_layout()
plt.show()
```

```python
# Step 8: Show differences between even and odd datasets
even_home_prices = home_prices[:100]  # Just as an example, use first 100 prices
mean_even = mean(even_home_prices)
sorted_even_prices = sorted(even_home_prices)
n_even = len(sorted_even_prices)
median_even = (sorted_even_prices[n_even // 2 - 1] + sorted_even_prices[n_even // 2]) / 2

try:
    mode_even = mode(even_home_prices)
except StatisticsError:
    mode_even = "No unique mode"
```

```python
# Print results for even dataset
print("\nEven Dataset (first 10):", even_home_prices[:10])  # Show only first 10 for brevity
print("Mean (Even):", mean_even)
print("Median (Even):", median_even)
print("Mode (Even):", mode_even)
```

```python
# Odd dataset (using the first 99 prices to ensure it's odd)
odd_home_prices = home_prices[:99]  # Use first 99 prices for odd dataset
mean_odd = mean(odd_home_prices)
sorted_odd_prices = sorted(odd_home_prices)
n_odd = len(sorted_odd_prices)
median_odd = sorted_odd_prices[n_odd // 2]  # For odd count, take the middle element
```

```python
try:
    mode_odd = mode(odd_home_prices)
except StatisticsError:
    mode_odd = "No unique mode"
```

```python
# Print results for odd dataset
print("\nOdd Dataset (first 10):", odd_home_prices[:10])  # Show only first 10 for brevity
print("Mean (Odd):", mean_odd)
print("Median (Odd):", median_odd)
print("Mode (Odd):", mode_odd)
```
